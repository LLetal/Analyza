# Omezenost (2.2.3) - nepovinné, ale radši si to sem píšu, používá se u dalších důkazů
Nechť A je množina s úplným uspořádáním (splňuje-li antisymetrii, tranzitivitu a reflexivitu - jinými slovy prostě třeba množina seřazená od 0 do 100) a máme jistou množinu $B\subset A$. Řekneme, že množina B je omezená shora, jestliže existuje její horní závora v A($\exists y \in A x \in B => x \leq y$). Řekneme, že množina B je omezená zdola, jestliže existuje nějaká dolní závora v A ($\exists y \in A x \in B => x \geq y$). Jestliže je množina omezená shora i zdola, tak řekneme, že je *omezená*

# Maximum a minimum (2.2.4)
Nechť máme množinu A s úplným uspořádáním $B \subset A$. Prvek M náležící B nazveme *maximem* právě tehdy, když je horní závorou této množiny ($\forall x \in B => \forall x\leq M$). Naproti tomu prvek m nazveme *minimem* množiny, pakliže je dolní závorou naší množiny.

Názorný příklad:
$$
B = [0,100], M = 100, m = 0
$$
Právě všechny prvky této množiny jsou totiž menší anebo rovny 100 a současně jsou všechny prvky větší, anebo rovny 0.


# Infimum a Supremum
Nechť A je množina s úplným uspořádáním (Prostě seřazené třebas od největší k největší, jednoduše je tam nějaký pattern) a nechť $B \subset A$. Jistý prvek S množiny A nazveme *supremem* právě tehdy, když je nejmenší horní závorou množiny B. Kvantifikátory řečeno platí:
$$
x \in B => x \leq S, (y \in A \land y < S)=> \exists x \in B => x > y
$$
Pozn.: Číslo y může být jakkoliv nepatrně menší, než S. Vznikne třebas tak, že od S odečteme jedničku. \

Naproti tomu prvek I množiny B nazveme jeho infimem, právě tehdy, když se jedná o největší možnou dolní závoru této množiny (opět musí náležet nadmnožině A). Kvantifikátory řečeno:
$$
\forall x \in B => x \geq I, y' \in A \land y' > I => \exists x' \in B => x' < y'
$$

# Spojitost funkce v bodě (3.2.1)
Mějme funkci $f:\mathbb{R}  \to \mathbb{R} $ a $x_0 \in \mathbb{R} $. Funkce f je potom v tomto bodě spojitá právě tehdy, když:
$$
\lim_{x\to x_0} f(x) = f(x_0)
$$
Spojitost zleva a zprava získáme analogicky jako u limity, jednoduše se bude zprava rovnat funkční hodnotě v pravém okolí bodu $x_0$
# Spojitost funkce na intervalu (6.2.13)
Mějme funkci $f:\mathbb{R}  \to \mathbb{R} $ a $x_0 \in \mathbb{R} $. Funkce f je potom na nějakém intervalu I spojitá právě tehdy, když:
$$
\forall \epsilon >0 \, \exists \delta > 0 \, x,y \in I \land 0<|x-y|< \delta => |f(x)-f(y)| < \epsilon
$$
Pozn.: Za zmínku stojí i stejnoměrná spojitost, jejíž definice se od definice spojitosti na intervalu liší jen zdánlivě nepratrně.:
$$
\forall \epsilon >0 \, \forall x \in I \, \exists \delta > 0 \, x,y \in I \land 0<|x-y|< \delta => |f(x)-f(y)| < \epsilon
$$
Tato "Malá změna" však říká, že pro všechna x existuje právě $\delta$, jenž je pro všechny x stejné. Slovy pana Bráni můžeme funkcí protáhnout nějakou trubku, aniž bychom jí zkracovali/jinak měnili její tvar (6.2.12).

# Derivace funkce (3.3.1)
Nechť $f: \mathbb{R}  \to \mathbb{R}, x \in \mathbb{R}$ a $A \in \mathbb{R}^*$. Řekneme, že funkce $f$ má v bodě $x_0$ *derivaci* rovnou A právě tehdy, když:
$$
\lim_{x\to x_0} f(x) = \frac{f(x)-f(x_0)}{x-x_0} = A
$$
Současně platí (podobně jako u limit a spojitosti), že derivace v bodě $x_0$ existuje, existuje-li derivace zprava i zleva. Pročež si je zadefinujme:
$$
\lim_{x\to x_0^{-}} f(x) = \frac{f(x)-f(x_0)}{x-x_0} = A
$$
Pro limitu zleva
Pozn.: ekvivalentní definice derivace - častěji se s ní stekáváme ve fyzice:
$$
\lim_{\Delta \to 0} f(x) =\frac{f(x_0 - \Delta)}{\Delta} =A
$$
kdybychom chtěli derivaci zleva a k tomu derivaci zprava, tak ekvivalentně přepíšeme tak, že dodáme znaménka u nuly.

# Primitivní funkce (4.1.1)
Nechť $f,F: \mathbb{R} \to \mathbb{R}$. Funkce F je primitivní funkcí k f na jistém otevřeném intervalu právě tehdy, když platí, že $F' = f$ na $(a,b)$. Pomocí hada značíme jako $\int f dx = F$.

Pozn.: Otevřený interval v definici vystupuje právě proto, aby zabrala co nejširší okruh funkcí. Kdybychom místo toho napsali uzavřený, tak existují funkce, které na uzavřením intervalu nemají primitivní funkce. Příklad.: $f(x) = \frac{1}{x}$. Tato funkce má primitivní funkci na intevalu $(0,\infty)$, ne už však na intervalu $[0, \infty)$, což plyne z definice logaritmu.

# Lokální a globární extrém funkce na množině (6.1.1)
Mějme $f: \mathbb{R} \to \mathbb{R}$ a $x_0 \in D_f$. Řekneme, že $f$ á v tomto bodě *lokální maximum*, existuje-li $\delta >0$ takové, že 
$$
f(x) \leq f(x_0), x \in U_\delta (x_0)
$$
O *ostrém lokálním maximu* hovoříme tehdy, když:

$$
f(x) < f(x_0), x \in P_\delta (x_0)
$$
Analogicky hovnoříme o *lokálním minimu*, existuje-li $\delta >0$ takové, že:
$$
f(x) \geq f(x_0), x \in U_\delta (x_0)
$$

$$
f(x) > f(x_0), x \in P_\delta (x_0)
$$
Všimněme si analogie s maximem a minimem množin. Prstencové okolí u ostrých maxim/minim zavádím proto, že vylučujeme samotný bod $x_0$ a dáváme najevo, že body v tomto okolí jsou od $x_0$ nějak vzdáleny. Globální minimum/maximum je poté totéž, ale vztažené na celou množinu $M$ definičního oboru, potažmu intervalu, na kterém funkci zkoumáme. 
# Monotonie funkce(6.4.1)
Nechť $f: \mathbb{R} \to \mathbb{R}$ a $x_0 \in \mathbb{R}$. Funkce f je rostoucí v bodě $x_0$, pokud existuje $\delta >0$ takové, že:
$$
f(x) < f(x_0), x \in P_\delta ^- (x_0) \land f(x) > 0 \, f(x_0) \, pro \, x \in P_\delta ^+ (x_0)
$$
Pro klesající potom platí opak.
Pozn.: Hovoříme-li o funkci monotónní na nějakém intervalu, tak platí že všechny funkční hodnoty na daném intervalu jsou uspořádané. 


# Konvexivita a konkávnost(6.5.1)
Nechť $f: \mathbb{R} \to  \mathbb{R}$ a máme nějaký interval I. Funkce f je *konvexní* právě tehdy, když  pro všechna $x,y,z \in I$ splňující $x< y<z$ platí:

$$
f(y) \leq f(x) + \frac{f(z)-f(x)}{z-x}(y-x)
$$
Pravá strana nerovnosti představuje výraz pro přímku z bodu funkční hodnoty v bodě x do bodu z. Je-li nerovnost ostrá říkáme, že je funkce *ryze konvexní*. Obrátíme-li znaménko nerovnost/rovnosti, získáme konkávní, případně ryze konkávní funkci na nějakém intervalu I. \

# Riemannův integrál 
Nechť $f$ a $[a,b]$ jsou jako výše. *Dolním Rieamannovým integrálem* nazveme číslo:
$$
(R) \int_{\overline a}^{b} f(x) dx := \sup_{D} s(f,D)
$$
*Horním Rieamannovým integrále* nazveme:
$$
(R) \int_{a}^{\overline b} f(x) dx = \inf_{D} S(f,D)
$$
Pokud platí, že  $(R)\int_{a}^{b} f(x) dx = (R)\int_{a}^{b} f(x) dx$, tuto společnou hodnotu nazveme *Riemannovým integrálem* a značíme ji:
$$
(R) \int_{a}^{b} f(x)dx
$$
## Alternativní (Braňova definice):
Funkce je Rieamannovsky integrovatelná, pokud $f:[a,b] \to \mathbb{R}$, pokud:
$$
\int_{b}^{\overline a} f(x) dx = \int_{\overline b}^{a} f(x) dx 
$$
